```python
import json
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

tokenized = []
file_path = "./aida_train_enriched8.jsonl"
from datasets import Dataset

original_data_aida = []
with open(file_path, "r") as file:
    for line in file:
        # Load each line as a JSON object
        data_line = json.loads(line)
        additional_question = "<additional> "
        try:
            for related in data_line["most_related"]:
                length = len(related)
                if len(related) == 4:
                    additional_question += related[2]
                    additional_question += " "
        except:
            None
        additional_question += "</additional>"
        data_line["question"] = " Not In Candidates </ec>" + data_line["question"]
        data_line["context"] += additional_question
        original_data_aida.append(data_line)
print(len(original_data_aida))
```

```python
import random
random.shuffle(original_data_aida)

to_nil = original_data_aida[:int(len(original_data_aida)*0.4)]
not_to_nil = original_data_aida[int(len(original_data_aida)*0.4):]
for item in to_nil:
    answer = item["answers"]['text']
    item["question"] = item["question"].replace(answer[0] + ' </ec>', "")
    item["question"] = " Not In Candidates </ec>" + item["question"]
    item["answers"]['text'][0] = "Not In Candidates"

concat = to_nil + not_to_nil

#write concat to jsonl file
with open("./aaida_nilled.jsonl", "w") as file:
    for item in concat:
        json.dump(item, file)
        file.write("\n")
```

```python
from tqdm import tqdm
import json
import torch
import random

results = []
file_path = "./msnbc-test-kilt.jsonl"
dataset = []
with open(file_path, "r") as file:
    for line in file:
        # Load each line as a JSON object
        data_line = json.loads(line)

        dataset.append(data_line)
random.shuffle(dataset)

to_nil = dataset[: int(len(dataset) * 0.7)]
not_to_nil = dataset[int(len(dataset) * 0.7) :]
for item in to_nil:
    try:
        print(item["output"][0]["answer"])
        item["candidates"].append("Not In Candidates")
        result = list(
            filter(
                lambda x: item["output"][0]["answer"]  not in x,
                item["candidates"],
            )
        )
        item["candidates"] = result
        item["output"][0]["answer"] = "Not In Candidates"
    except Exception as e:
        print(e)

concat = to_nil + not_to_nil

# write concat to jsonl file
with open("./msnbc_nilled.jsonl", "w") as file:
    for item in concat:
        json.dump(item, file)
        file.write("\n")
```
