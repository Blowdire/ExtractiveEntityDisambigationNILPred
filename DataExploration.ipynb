## Aida train

```python
import json

dataset = []
inputs = ''
with open('./aida-train-kilt.jsonl', "r") as file:
    for line in file:
        # Load each line as a JSON object
        data_line = json.loads(line)
        data_line['input'] = data_line['input'].replace('[START_ENT]', '').replace('[END_ENT]', '')
        inputs += data_line['input'] + ' '
        dataset.append(data_line)

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a word cloud
wordcloud = WordCloud(font_path="./AovelSansRounded-rdDL.ttf").generate(inputs)

# Display the word cloud
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import pandas as pd

df = pd.read_json("./aida-train-kilt.jsonl", lines=True)
df["answer"] = df["output"].apply(lambda x: x[0]["answer"])
```

```python
mention_frequency = df.groupby("answer").size()
```

```python
import matplotlib.pyplot as plt

plt.tick_params(
    axis="x",  # changes apply to the x-axis
    which="both",  # both major and minor ticks are affected
    bottom=False,  # ticks along the bottom edge are off
    top=False,  # ticks along the top edge are off
    labelbottom=False,
)
plt.ylim(0, 80)
plt.xlabel("Mentions")
plt.ylabel("Mention Frequency")
plt.title("Mention Frequency of AIDA Train Dataset")
mention_frequency_sorted = mention_frequency.sort_values(ascending=False)
mention_frequency_sorted.plot.bar()
plt.show()
```

```python
import numpy as np

length = 0
lengths = np.array([])
for item in dataset:
    length += len(item["candidates"])
    lengths = np.append(lengths, len(item["candidates"]))

print(f"The mean length of the candidates is {length/len(dataset)}")
print(f"The median length of the candidates is {np.median(lengths)}")
print(f"The mode length of the candidates is {np.median(lengths)}")
```

```python
words = inputs.split()

# Use a set to get the unique words
unique_words = set(words)

# The number of unique words is the size of the set
num_unique_words = len(unique_words)

print(f"The number of unique words is {num_unique_words}.")
```

# correct candidate position

```python
import pandas as pd
from tqdm import tqdm

positions = []

datasets = ["aida", "msnbc", "aquaint", "ace2004", "clueweb", "wiki"]

for dataset in tqdm(datasets):
    dsPath = (
        f"./Datasets/Base/{dataset}-test-kilt.jsonl"
        if dataset != "aida"
        else f"./Datasets/Base/{dataset}-train-kilt.jsonl"
    )
    df = pd.read_json(dsPath, lines=True)
    for index, item in df.iterrows():
        answer = item["output"][0]["answer"]
        candidates = item["candidates"]
        for i in range(len(candidates)):
            if candidates[i] == answer:
                positions.append(i)
                break
```

```python
positions[:5]
```

```python
import statistics

mode = statistics.mode(positions)
mean = statistics.mean(positions)
median = statistics.median(positions)
print(mode, mean, median)
```
